```python
import pandas as pd
import matplotlib.pyplot as plt
from itertools import cycle

# Reading the CSV file into a pandas DataFrame
file_path = 'yearlyadded.csv'
df = pd.read_csv(file_path)

# Getting unique crime types and assign a color to each type
crime_types = df['CrimeType'].unique()
colors = cycle(plt.cm.tab10.colors)

color_mapping = {crime_type: next(colors) for crime_type in crime_types}

# Create a bar chart for each crime type
plt.figure(figsize=(10, 6))

for crime_type, color in color_mapping.items():
    crime_data = df[df['CrimeType'] == crime_type]
    plt.bar(crime_data['Year'].astype(str), crime_data['Count'], label=crime_type, color=color)

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Complaints Crime Counts Over Years')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the CSV file into a pandas DataFrame
file_path = 'yearlyadded.csv'
df = pd.read_csv(file_path)

# Filtering data for the crime type 'Administrative Code'
admin_code_data = df[df['CrimeType'] == 'ADMINISTRATIVE CODE']

# Creating a line graph for the Administrative Code over years
plt.figure(figsize=(10, 6))

plt.plot(admin_code_data['Year'], admin_code_data['Count'], marker='o', linestyle='-', color='b')

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Comnplaints Administrative Code Crime Over Years')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the data from the CSV file
df = pd.read_csv('age.csv')

# Grouping the DataFrame by Year and sum the OffenseCount for each year
df_total_crimes = df.groupby('Year')['OffenseCount'].sum().reset_index()

# Plotting the line graph for all crimes over the years
plt.figure(figsize=(10, 6))
plt.plot(df_total_crimes['Year'], df_total_crimes['OffenseCount'], marker='o', linestyle='-', color='b')
plt.title('Total Crimes Committed Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Offense Count')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading data from the CSV file
df = pd.read_csv('age.csv')

# Grouping the DataFrame by Year and Age Group, and sum the OffenseCount for each group
df_grouped = df.groupby(['Year', 'VIC_AGE_GROUP'])['OffenseCount'].sum().reset_index()

# Creating a separate line for each age group
age_groups = df_grouped['VIC_AGE_GROUP'].unique()

plt.figure(figsize=(12, 8))

for age_group in age_groups:
    df_age_group = df_grouped[df_grouped['VIC_AGE_GROUP'] == age_group]
    plt.plot(df_age_group['Year'], df_age_group['OffenseCount'], label=f'Age {age_group}')

plt.title('Crimes Committed Over the Years by Age Group')
plt.xlabel('Year')
plt.ylabel('Offense Count')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading data from the CSV file
df = pd.read_csv('yearlyadded.csv')

# Pivoting the DataFrame to have 'Year' as index, 'OffenseType' as columns, and 'OffenseCount' as values
df_pivot = df.pivot(index='Year', columns='CrimeType', values='Count').fillna(0)

# Plotting the line graph for different crime types over the years
plt.figure(figsize=(14, 8))

for crime_type in df['CrimeType'].unique():
    plt.plot(df_pivot.index, df_pivot[crime_type], label=crime_type)

plt.title('Crimes Committed Over the Years by Crime Type')
plt.xlabel('Year')
plt.ylabel('Offense Count')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading data from the CSV file
df = pd.read_csv('sex.csv')

# Pivoting the DataFrame to have 'Year' as index, 'VIC_SEX' as columns, and 'OffenseCount' as values
df_pivot = df.pivot(index='Year', columns='VIC_SEX', values='OffenseCount').fillna(0)

# Plotting the line graph for different victim sexes over the years
plt.figure(figsize=(10, 6))

for victim_sex in df['VIC_SEX'].unique():
    plt.plot(df_pivot.index, df_pivot[victim_sex], label=victim_sex)

plt.title('Crimes Committed Over the Years by Victim Sex')
plt.xlabel('Year')
plt.ylabel('Offense Count')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
